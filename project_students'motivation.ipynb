## Comparaison des motivations IND des étudiants d'origine immigrés de 1e VS 2e génération

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Lecture des données à partir du fichier Excel
data = pd.read_excel('data/IND EDI 1 VS 2.xlsx')
```

```python
# Séparation des données en deux variables : première génération et seconde génération
premiere_generation = data["IND EDI 1G"].values.reshape(-1, 1)
seconde_generation = data["IND EDI 2G"].values.reshape(-1, 1)
```

```python
# Tracer un graphique de dispersion
plt.scatter(premiere_generation, seconde_generation, color='blue', label='Données')
```

```python
import numpy as np

# Remplacer les valeurs manquantes par la moyenne
premiere_generation = np.nan_to_num(premiere_generation, nan=np.nanmean(premiere_generation))
seconde_generation = np.nan_to_num(seconde_generation, nan=np.nanmean(seconde_generation))

# Construction du modèle de régression linéaire
model = LinearRegression()
model.fit(premiere_generation, seconde_generation)
```

```python
# Construction du modèle de régression linéaire
model = LinearRegression()
model.fit(premiere_generation, seconde_generation)
```

```python
# Tracer la ligne de régression
plt.plot(premiere_generation, model.predict(premiere_generation), color='red', label='Régression linéaire')

```

```python
# Ajouter des labels et une légende
plt.xlabel('Motivations Première Génération')
plt.ylabel('Motivations Seconde Génération')
plt.title('Comparaison des motivations entre première et seconde génération')

# Spécifier les artistes à inclure dans la légende
plt.legend(['Données', 'Régression linéaire'])

# Affichage du graphique
plt.show()
```

```python
import seaborn as sns

# Tracer un histogramme de distribution des motivations pour les étudiants de première génération
plt.title('Distribution des motivations pour les étudiants de première génération')
sns.distplot(premiere_generation)
plt.xlabel('Motivations')
plt.show()

# Tracer un histogramme de distribution des motivations pour les étudiants de seconde génération
plt.title('Distribution des motivations pour les étudiants de seconde génération')
sns.distplot(seconde_generation)
plt.xlabel('Motivations')
plt.show()
```

```python
 import scipy.stats as stats

# Comparer les moyennes avec le test t-student
t_statistic, p_value = stats.ttest_ind(premiere_generation, seconde_generation)
print("Test t-student - Statistique de test :", t_statistic)
print("Test t-student - Valeur de p :", p_value)

# Visualiser les distributions
plt.title('Distribution des motivations')
sns.distplot(premiere_generation, label='Première Génération')
sns.distplot(seconde_generation, label='Seconde Génération')
plt.xlabel('Motivations')
plt.legend()
plt.show()

# Calculer d'autres mesures de tendance centrale
mean_premiere_generation = np.mean(premiere_generation)
median_premiere_generation = np.median(premiere_generation)
mode_premiere_generation = stats.mode(premiere_generation)

mean_seconde_generation = np.mean(seconde_generation)
median_seconde_generation = np.median(seconde_generation)
mode_seconde_generation = stats.mode(seconde_generation)

print("Moyenne première génération :", mean_premiere_generation)
print("Médiane première génération :", median_premiere_generation)
print("Mode première génération :", mode_premiere_generation.mode[0])

print("Moyenne seconde génération :", mean_seconde_generation)
print("Médiane seconde génération :", median_seconde_generation)
print("Mode seconde génération :", mode_seconde_generation.mode[0])
```
